```python
import pandas as pd
pd.options.mode.chained_assignment = None
import csv
import numpy as np

from google_play_scraper import Sort, reviews, reviews_all
from itertools import count
```

# 2. Scarping Data dari Playstore App Cici: Asisten AI Cerdas

---
Aplikasi Cici merupakan aplikasi asisten AI yang dapat digunakan untuk pekerjaan, belajar, dan mencari ide kreatif.

---
Melakukan scraping data pada review aplikasi Cici: Asisten AI Cerdas di playstore

```python
result, continual_token = reviews(
    'com.larus.wolf',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=12000,
    filter_score_with=None
)

df = pd.DataFrame(np.array(result),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df
```

```python
df.info()
```

# 3. Cleaning Dataset

Melakukan beberapa metode pembersihan sederhana untuk dataset scraping

```python
new_df = df[['reviewId', 'userName', 'score', 'content', 'thumbsUpCount', 'at']]
new_df.info()
```

```python
clean_df = new_df.dropna()
clean_df = clean_df.drop_duplicates()
clean_df.info()
```

# 4. Saving Dataset to CSV

```python
clean_df.to_csv('cici_reviews.csv', index=False)
```
